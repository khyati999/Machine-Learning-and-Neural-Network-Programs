```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
california_housing = fetch_california_housing(as_frame=True)
data=california_housing.frame

data.head()

```

```python
y=data['MedHouseVal']       # Target var
X=data.drop(['MedHouseVal'],axis=1)             # Feature var
```

```python
# .T transposes the results, transforming rows into columns
X.describe().T
```

```python
from sklearn.model_selection import train_test_split 
SEED=42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED) #this function 
```

```python
print(len(X))       # 20640
print(len(X_train)) # 15480
print(len(X_test))  # 5160
```

```python
from sklearn.preprocessing import StandardScaler
# Fit only on X_train
scaler = StandardScaler()
scaler.fit(X_train)
# Scale both X_train and X_test
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
col_names=['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']
scaled_df = pd.DataFrame(X_train, columns=col_names)
scaled_df.describe().T
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors=12)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'mae: {mae}')
print(f'mse: {mse}')
print(f'rmse: {rmse}')
```

```python
regressor.score(X_test, y_test)
```
